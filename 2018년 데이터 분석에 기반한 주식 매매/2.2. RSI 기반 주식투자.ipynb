```python
sql = ''
sql = '''
SELECT T1.STK_CD, T1.DT
      , T1.C_PRC, T1.O_PRC, T1.H_PRC, T1.L_PRC, T1.VOL
FROM history_dt T1
WHERE T1.STK_CD = 121800
AND T1.DT >= '20190101'
AND T1.DT < '20191231'
;
'''
```

```python
dt_df = pd.DataFrame(res)
dt_df
```

```python
dt_df['DT'] = pd.to_datetime(dt_df['DT'])
```

```python
dt_df.loc[:, 'C_PRC' : ] = dt_df.loc[:, 'C_PRC' : ].astype(int)
```

```python
dt_df.info()
```

```python
dt_df['DIFF'] = dt_df['C_PRC'].diff()
```

```python
def UD(diff):
  ups = 0
  downs = 0
  if diff > 0:
    ups = diff
  elif diff < 0:
    downs = abs(diff)
  return pd.Series([ups, downs])
```

```python
dt_df[['U', 'D']] = dt_df['DIFF'].apply(UD)
```

```python
dt_df['AU'] = dt_df['U'].rolling(14).mean()
```

```python
dt_df['AD'] = dt_df['D'].rolling(14).mean()
```

```python
dt_df['RS'] = dt_df['AU'] / dt_df['AD']
```

```python
dt_df['RSI'] = dt_df['RS'] / (1 + dt_df['RS'])
```

```python
dt_df['RSI_signal'] = dt_df['RSI'].rolling(6).mean()
```

```python
dt_df = dt_df.set_index('DT')
```

```python
dt_df['S3'] = 0.3
dt_df['S7'] = 0.7
```

```python
dt_df.reset_index(inplace=True)
```
