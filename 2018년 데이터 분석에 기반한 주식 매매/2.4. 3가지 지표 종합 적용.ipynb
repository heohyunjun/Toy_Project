# 비덴트 종합지표 적용 시각화

```python
sql = ''
sql = '''
SELECT T1.STK_CD, T1.DT
      , T1.C_PRC, T1.O_PRC, T1.H_PRC, T1.L_PRC, T1.VOL
FROM history_dt T1
WHERE T1.STK_CD = 121800
AND T1.DT >= '20190101'
AND T1.DT < '20191231'
;
'''
```

```python
cursor.execute(sql)
res = cursor.fetchall()
res
```

```python
dt_df = pd.DataFrame(res)
dt_df
```

```python
dt_df.info()
```

```python
dt_df['DT'] = pd.to_datetime(dt_df['DT'])
```

```python
dt_df.info()
```

```python
dt_df.loc[:, 'C_PRC' : ] = dt_df.loc[:, 'C_PRC' : ].astype(int)
```

```python
dt_df.info()
```

```python
fig = go.Figure()

fig.add_trace(
    go.Scatter(x=dt_df['DT'] # 2019년 주식장 열린 모든 일자
              , y=dt_df['C_PRC'] # 종가
              , mode='lines'
              , name='비텐트 종가 선그래프'
    )
)

fig.add_trace(
    go.Scatter(x=['2019-02-11'] # 매수일
              , y=[5750.000] # 매수일의 종가
              , mode='markers+text'
              , text='2/11 5,750원'
              , marker = dict(color = 'red', size = 10, symbol = 17)
              , textfont = dict(size = 15)
              , textposition = "bottom center"
              , name='지표 종합적용 매수일'
    )
)

fig.add_trace(
    go.Scatter(x=['2019-05-17'] # 매도일
              , y=[6620] # 매수일의 종가
              , mode='markers+text'
              , text='5/17 6,620원'
              , marker = dict(color = 'blue', size = 10, symbol = 17)
              , textfont = dict(size = 15)
              , textposition = "top center"
              , name='지표 종합적용 매도일'
    )
)

fig.show()
```
